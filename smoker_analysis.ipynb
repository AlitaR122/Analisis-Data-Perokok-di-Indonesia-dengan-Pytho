# 1. Mengurutkan berdasarkan umur

```python
pivot_smoker = smoker.pivot(index='Umur', columns='Tahun', values='Persentase Perokok')
pivot_smoker
```

# 2. Nilai tertinggi persentase perokok

```python
smoker_max = smoker.loc[smoker.groupby('Tahun')['Persentase Perokok'].idxmax()]
smoker_max
```

# 3. Nilai terrendah persentase perokok

```python
smoker_min = smoker.loc[smoker.groupby('Tahun')['Persentase Perokok'].idxmin()]
smoker_min
```

# 4. Nilai rata-rata persentase perokok berdasarkan tahun

```python
smoke_mean = smoker.groupby('Tahun')['Persentase Perokok'].mean()
smoke_mean
```

# 5. Nilai rata-rata persentase perokok berdasarkan umur

```python
smoke_umur_mean = smoker.groupby('Umur')['Persentase Perokok'].mean().sort_values(ascending=False).tail(10)
smoke_umur_mean
```

# 6. Jumlah Persentase perokok berdasarkan tahun

```python
smoker_tahun_sum = smoker.groupby('Tahun')['Persentase Perokok'].sum().reset_index()
smoker_tahun_sum
```

```python
smoker_umur_sum = smoker.groupby('Umur')['Persentase Perokok'].sum().reset_index()
pd.options.display.max_rows = 15
smoker_umur_sum
```
